```python
test = 0
for i in data['name']:
    for j in range(1980, 2017):
        if str(j) in str(i):
            test += 1         
test
```

```python
data['isna_year_of_release'] = data['year_of_release'].isna().astype(int)
name_release = data['name'].unique().tolist()
name_years_dict = {}
for i in range(1980, 2017):
    for j in name_release:
        if str(i) in str(j):
            name_years_dict.update({j: i})

def isna_year(row):
    if row['isna_year_of_release'] == 1 and row['name'] in name_years_dict.keys():
        return name_years_dict[row['name']]
    else:
        return row['year_of_release']


data['year_of_release'] = data.apply(isna_year, axis=1)
data['year_of_release'].isna().sum()
```

```python
def new_rating(value):
    if value == 'E':
        return 6
    elif value == 'M':
        return 17
    elif value == 'T':
        return 13
    elif value == 'E10+':
        return 10
    elif value == 'K-A':
        return 6
    elif value == 'AO':
        return 18
    elif value == 'EC':
        return 3
    elif value == 'RP':
        return 1 # –æ–∂–∏–¥–∞–µ—Ç –æ—Ü–µ–Ω–∫–∏
    else:
        return 100 # –Ω–µ—Ç –æ—Ü–µ–Ω–∫–∏

data['new_rating'] = data['rating'].apply(new_rating)
```

–°–æ–æ—Ç–Ω–∏—Å—Ç–∏ –≤—ã–≤–æ–¥—ã —Å –ø—Ä–æ–¥–∞–∂–∞–º–∏ –Ω–∞ –¥—Ä—É–≥–∏—Ö –ø–ª–∞—Ç—Ñ–æ—Ä–º–∞—Ö

```python

def score_sales(platform):  
    global platforms_more_sales
    platforms_sales = platforms_more_sales.loc[platforms_more_sales['platform'].isin(['PS4','XOne', 'PSV', 'PC', '3DS'])].copy()
    score_sales_axemple = platforms_sales.loc[platforms_more_sales['platform'] == platform][['all_sales', 'critic_score', 'user_score']].corr()
    X360 = platforms_sales.loc[platforms_sales['platform'] == platform][['platform','all_sales', 'critic_score', 'user_score']]

    plat_name = X360['platform'].drop_duplicates().tolist()

    plt.figure(figsize=(5, 3))
    ax = sns.heatmap(score_sales_axemple, annot=True);
    ax.set_title(f'–ø–ª–∞—Ç—Ñ–æ—Ä–º–∞ {plat_name[0]}');

    plt.figure(figsize=(5, 3))
    ax = sns.scatterplot(y='critic_score', x='all_sales', hue='critic_score', data=X360);
    ax.set_title(f'–ø–ª–∞—Ç—Ñ–æ—Ä–º–∞ {plat_name[0]}');

    plt.figure(figsize=(5, 3))
    ax = sns.scatterplot(y='user_score', x='all_sales', hue='user_score', data=X360);
    ax.set_title(f'–ø–ª–∞—Ç—Ñ–æ—Ä–º–∞ {plat_name[0]}');

score_sales(platform='XOne')    
```

```python
def portrait_user(area):
    if area == 'na_sales':
        name_area = '—Å–µ–≤–µ—Ä–Ω–∞—è –ê–º–µ—Ä–∏–∫–∞'
        
    elif area == 'eu_sales':
        name_area = '–ï–≤—Ä–æ–ø–∞'
        
    elif area == 'jp_sales':
        name_area = '–Ø–ø–æ–Ω–∏—è'
        
    elif area == 'all_sales':
        name_area = '–≤—Å–µ —Ä–µ–≥–∏–æ–Ω—ã'
        
    else:
        name_area = '–¥—Ä—É–≥–∏–µ —Ä–µ–≥–∏–æ–Ω—ã'
        
    fig, axis = plt.subplots(1,3, figsize=(14,7))
    plt.subplots_adjust(wspace=0.5, hspace=0.5)
    


    user_portrait_na = user_portrait.groupby('platform')[area].agg('sum').sort_values(ascending=False).reset_index()
    
    head_len = 0
    for i in range(5):
        head_len += 1
        if user_portrait_na.loc[i][area] < 6:
                    head_len = i 
    num = len(user_portrait_na) - len(user_portrait_na.head(head_len))
    other = pd.DataFrame({'platform': 'other',
                         area: [user_portrait_na[area].tail(num).sum()]})
   
    user_portrait_na = pd.concat([user_portrait_na, other], ignore_index=True).sort_values(by=area,ascending=False)
    user_portrait_na = user_portrait_na.head(head_len + 1)
    
    user_portrait_na = (user_portrait_na.plot.pie(labels = user_portrait_na['platform'], 
                         autopct='%1.2f%%', y=area, legend=False, ax=axis[0])
                        .set_title(f'–ø–ª–∞—Ç—Ñ–æ—Ä–º—ã {name_area} '));
   
    
    
    user_portrait_na = user_portrait.groupby('genre')[area].agg('sum').sort_values(ascending=False).reset_index()
    num = (len(user_portrait_na) - head_len)
    
    other = pd.DataFrame({'genre': 'other',
                         area: [user_portrait_na[area].tail(num).sum()]})
    
    user_portrait_na = pd.concat([user_portrait_na, other], ignore_index=True).sort_values(by=area,ascending=False)
    user_portrait_na = user_portrait_na.head(head_len + 1)
    
    user_portrait_na = (user_portrait_na.plot.pie(labels = user_portrait_na['genre'], 
                         autopct='%1.2f%%', y=area, legend=False, ax=axis[1])
                        .set_title(f'–∂–∞–Ω—Ä—ã {name_area} '));
    
    
    
    user_portrait_na = user_portrait.groupby('new_rating')[area].agg('sum').sort_values(ascending=False).reset_index()
    num = (len(user_portrait_na) - head_len)
    
    other = pd.DataFrame({'new_rating': 'other',
                         area: [user_portrait_na[area].tail(num).sum()]})
    
    user_portrait_na = pd.concat([user_portrait_na, other], ignore_index=True).sort_values(by=area,ascending=False)
    user_portrait_na = user_portrait_na.head(head_len)
    
    user_portrait_na = (user_portrait_na.plot.pie(labels = user_portrait_na['new_rating'], 
                         autopct='%1.2f%%', y=area, legend=False, ax=axis[2])
                        .set_title(f'—Ä–µ–π—Ç–∏–Ω–≥ {name_area}'));
    
portrait_user(area='na_sales')
```

## –ü—Ä–æ–≤–µ—Ä–∏—Ç—å –≥–∏–ø–æ—Ç–µ–∑—ã

#### –°—Ä–µ–¥–Ω–∏–µ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—å—Å–∫–∏–µ —Ä–µ–π—Ç–∏–Ω–≥–∏ –ø–ª–∞—Ç—Ñ–æ—Ä–º Xbox One –∏ PC –æ–¥–∏–Ω–∞–∫–æ–≤—ã–µ
–ü–æ—Ä–æ–≥–æ–≤–æ–µ –∑–Ω–∞—á–µ–Ω–∏–µ alpha –æ–ø—Ä–µ–¥–∏–ª–∏—Ç—å —Å–∞–º–æ—Å—Ç–æ—è—Ç–µ–ª—å–Ω–æ

<font color='DarkGreen'>üëå –•–æ—Ä–æ—à–æ, –ø—Ä–∞–≤–∏–ª—å–Ω–æ —Å—Ñ–æ—Ä–º—É–ª–∏—Ä–æ–≤–∞–Ω—ã –≥–∏–ø–æ—Ç–µ–∑—ã.</font>

```python
h0_hypothesis = data.loc[data['year'] >= 2012][['platform', 'user_score']].copy()
h0_hypothesis['isna_user_score'] = h0_hypothesis['user_score'].isna().astype(int)

Xbox_score_list = (h0_hypothesis.loc[(h0_hypothesis['platform'] == 'XOne') & 
                                     (h0_hypothesis['isna_user_score'] == 0)]['user_score'].to_list())
PC_score_list = (h0_hypothesis.loc[(h0_hypothesis['platform'] == 'PC') & 
                                   (h0_hypothesis['isna_user_score'] == 0)]['user_score'].to_list())
alpha = .05

results = st.ttest_ind(
    Xbox_score_list, 
    PC_score_list,
    equal_var = False) # –í–æ –≤—Ä–µ–º—è –≤—ç–±–∏–Ω–∞—Ä–∞ —Å–ø—Ä–∞—à–∏–≤–∞–ª–∏ —É –ì–ª–µ–±–∞ –ú–∏—Ö–∞–π–ª–æ–≤–∞ –µ–≥–æ —Ä–µ–∫–æ–º–µ–Ω–¥–∞—Ü–∏—è
                            # c—á–∏—Ç–∞—Ç—å –¥–∏—Å–ø–µ—Ä—Å–∏—é –≤—ã–±–æ—Ä–æ–∫ —Ä–∞–∑–Ω–æ–π 

    
print('P-value {:,.0%}'.format(results.pvalue))

if (results.pvalue < alpha):
    print(f'–û—Ç–≤–µ—Ä–≥–∞–µ–º –Ω—É–ª–µ–≤—É—é –≥–∏–ø–æ—Ç–µ–∑—É, —Ä–µ–π—Ç–∏–Ω–≥–∏ –ø–ª–∞—Ç—Ñ–æ—Ä–º Xbox –∏ PC —Ä–∞–∑–Ω—ã–µ')
else:
    print(f'P-value –±–æ–ª—å—à–µ –∑–∞–¥–∞–Ω–Ω–æ–π alpha, —Ä–µ–π—Ç–∏–Ω–≥–∏ –ø–ª–∞—Ç—Ñ–æ—Ä–º Xbox One –∏ PC –æ–¥–∏–Ω–∞–∫–æ–≤—ã–µ')
    print(f'H0 –≥–∏–ø–æ—Ç–µ–∑–∞ –≤–µ—Ä–Ω–∞')
```
